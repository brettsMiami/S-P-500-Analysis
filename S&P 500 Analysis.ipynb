```python
import pandas as pd
import numpy as np
```

```python
df2 = pd.read_csv('sp50.csv')
df2.set_index('No.')
```

```python
df = pd.read_csv('sp500.csv')
df.set_index('No.')
```

```python
df3 = pd.merge(df2,df, on = 'No.')
```

```python
df3.set_index('No.')
```

```python
df3 = df3.rename(columns= {'Ticker_x':'Ticker','Market Cap_x': 'Market Cap','Return on Assets': 'ROA','Return on Equity':'ROE',"Price_x":'Price','Change_x':'Change','Volume_x':'Volume',})
```

```python
df3.pop('Ticker_y')
df3.pop('Change_y')
df3.pop('Volume_y')
df3.pop('Market Cap_y')
```

```python
df3
```

```python
df3['ROE']= df3.ROE.replace({'%':''}, regex= True)
df3['ROA']= df3.ROA.replace({'%':''}, regex = True)
df3['ROE']=pd.to_numeric(df3.ROE)
df3['ROA']= pd.to_numeric(df3.ROA)
```

```python
df4 = df3.groupby('Sector')['ROE','ROA'].median()
df4
```

```python
df4.ROE.mean()
```

```python
df3.set_index('Sector')
Health= df3.loc[df3.Sector == 'Healthcare']
Health[ Health.ROE > 18.5].sort_values(by = 'Current Ratio',ascending = False)
```

```python
Consumer_defensive = df3.loc[df3.Sector == 'Consumer Defensive']
Consumer_defensive[Consumer_defensive.ROE > 43].sort_values('ROE', ascending = True)
```
